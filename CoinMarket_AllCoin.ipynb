```python
"""
Author:Sunil Kumar
Date:13 June,2018
This Script Gives a datafram of all the Currency inlisted on coinmarketcap with their initial launching price
"""

import requests
import time
import pandas as pd
import datetime
from datetime import timedelta
from bs4 import BeautifulSoup
import numpy as np
date=datetime.datetime(2013,4,28,0,0)
a=1
b=269 
d1 = {}
def month(a):
    if a < 10:
        return "0" + str(date.month)
    else:
        return str(date.month)
def day(a):
    if a < 10:
        return "0" + str(date.day)
    else:
        return str(date.day)
while a<b:            
    new=str(date.year)+month(date.month)+day(date.day)
    #     print('https://coinmarketcap.com/historical/'+new+'/')
    
    page=requests.get('https://coinmarketcap.com/historical/'+new+'/')
    soup=BeautifulSoup(page.content,'lxml')
    container=soup.find('table')
    coin=list(container.find_all('a',class_='currency-name-container'))
    price=list(container.find_all('a',class_='price'))
    
    for i in range(len(coin)):
        if coin[i].text not in d1.keys():
            d1[coin[i].text] = price[i].text
#     time.sleep()
    date=date+datetime.timedelta(days=7)
    a+=1
```

```python
"""
Author:Sunil Kumar
Date:13 June,2018
This Script Gives a datafram of all the Currency inlisted on coinmarketcap with their Current Price
"""
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
page=requests.get('https://coinmarketcap.com/all/views/all/')
soup=BeautifulSoup(page.content,'lxml')
container=soup.find('table')
coin_list1=list(container.find_all('a',class_='currency-name-container link-secondary'))
price_list1=list(container.find_all('a',class_='price'))
d2={}
for i in range(len(coin_list1)):
    if coin_list1[i].text not in d2.keys():
        d2[coin_list1[i].text] = price_list1[i].text
df2 = pd.DataFrame(index=None)        
df2["currency"], df2["price"] = d2.keys(), d2.values()
list_new=[]
Old_price=[]
for i in d2.keys():
    if i in d1.keys():
        list_new.append(i)
d3={'Currency':list_new,
    'Old_Price':[],
    'Current_Price':[],
    'Increase':[]}
for i in list_new:
    d3['Old_Price'].append(d1[i])
    d3['Current_Price'].append(d2[i])
df3.to_csv('Compare.csv',index=False)
```
